```python
""" Importing dependencies """
import matplotlib.pyplot as plt 
import pandas as pd 
import os
import preprocessor as p

from wordcloud import WordCloud, STOPWORDS
```

```python
stopwords = set(STOPWORDS) 
```

```python
def get_tweets(file):
    f = open(file, 'r', encoding='utf-8')
    tweets = []
    tweet_info = f.read().split("@" * 20)
    tweet_text = [tweet_info_single.split(">" * 10)[0].strip() for tweet_info_single in tweet_info]
    return tweet_text
```

```python
""" Fetch tweets for each region """

def get_term_wise_tweets(base_path):
    tweets_collection_term = {}
    for searchTerm in os.listdir(base_path):
        print(searchTerm)
        tweets_collection_term[searchTerm] = {}
        for city_file in os.listdir(os.path.join(base_path, searchTerm)):
            city = city_file[:-4]
            file_path = os.path.join(base_path, searchTerm, city_file)
            tweets = get_tweets(file_path)
            tweets_collection_term[searchTerm][city] = tweets
    return tweets_collection_term
```

```python
""" Some basic details """

base_path = os.path.join(os.pardir, "dataset", "tweets")
tweets_collection_term = get_term_wise_tweets(base_path)
```

```python
""" Getting the tweets for each hastag across the various cities """
term_wise_tweets = {}

for term, values in tweets_collection_term.items():
    temp = []
    for city, vals in tweets_collection_term[term].items():
        temp = temp+vals
        
    term_wise_tweets[term] = temp
```

```python
""" Creating the wordcloud for each hashtag """

def create_wordcloud(hashtag, word_list):
    wordcloud = WordCloud(width = 1000, height = 1000, 
                    stopwords = stopwords, 
                    background_color = 'white',
                    min_font_size = 10).generate(word_list) 

    # plot the WordCloud image                        
    plt.figure(figsize = (8, 8), facecolor = None) 
    plt.imshow(wordcloud)  
    plt.show() 
    plt.savefig('../results/'+hashtag+'_wordcloud')
    plt.close()
```

```python
for term, vals in term_wise_tweets.items():
    text = ""
    print(term)
    for tweet in vals:
        text = text + " " + p.clean(tweet)
        
    create_wordcloud(term, text)
```
